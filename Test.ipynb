```python
import pandas as pd
```

```python
data = pd.read_csv('ML_FOR POSTS1_base64_V1 (1).csv')
```

```python
data.head()
```

```python
# Dataset cleansing
```

```python
# Remove null image urls rows 
data = data[data['Image URL'].notna()]
```

```python
# Clean category columns
data['main category'] = data['main category'].str.replace('\t','')
data['Cat option 1'] = data['Cat option 1'].str.replace('\t','')
data['Cat option 2'] = data['Cat option 2'].str.replace('\t','')
data['Cat option 3'] = data['Cat option 3'].str.replace('\t','')
data['Cat option 4'] = data['Cat option 4'].str.replace('\t','')
```

```python
# Inserting category names into a list
cat = data['main category'].tolist()
cat1 = data['Cat option 1'].tolist()
cat2 = data['Cat option 2'].tolist()
cat3 = data['Cat option 3'].tolist()
cat4 = data['Cat option 4'].tolist()
```

```python
# Combine 4 category lists
cat = cat + cat1 + cat2 + cat3 + cat4
```

```python
# Removing NaN values
def get_cleaned_categories(category_list):
    cleansed_categories = []
    for i in category_list:
        if (type(i)==str):
            cleansed_categories.append(i.replace('\t', ''))
        else: pass
    
    return cleansed_categories
```

```python
cleansed_categories = get_cleaned_categories(cat)
```

```python
# Lowercasing categories
cleansed_categories = [x.lower() for x in cleansed_categories]
```

```python
# Getting final unique sorted category list
category_list = sorted(list(set(cleansed_categories)))
```

```python
# Use one hot encoding method to create vectors
```

```python
# Combining categories columns into one
data['categories'] = data[['main category', 'Cat option 1', 'Cat option 2', 'Cat option 3', 'Cat option 4']].values.tolist()
```

```python
data.head()
```

```python
# One hot encoding
one_hot_data = data['categories'].str.join('|').str.get_dummies()
```

```python
import requests

def download_image(image_name,pic_url):
    with open('data/{}.jpeg'.format(image_name), 'wb') as handle:
        response = requests.get(pic_url, stream=True)

        if not response.ok:
            print(response)

        for block in response.iter_content(1024):
            if not block:
                break

            handle.write(block)
```

```python
image_urls = data['Image URL']
image_index = 0
```

```python
def download_images(image_urls,image_index):
    for image_url in image_urls:
        # Download the image
        download_image(image_index,image_url)
    
        # Increment index
        image_index = image_index + 1
        
```
