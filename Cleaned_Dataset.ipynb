```python
import pandas as pd
```

```python
data = pd.read_csv('data/cleaned_data.csv')
```

```python
# Drop rows for url not working
```

```python
url_working_data = data[data['url working'] == True]
```

```python
# Drop unwanted columns
url_working_data = url_working_data[["Image URL","main category","Cat option 1","Cat option 2","Cat option 3","Cat option 4"]]
```

```python
url_working_data
```

```python
n = len(url_working_data)
```

```python
n
```

```python
# Create index list
index = [*range(0, n, 1)] 
```

```python
# Adding image index to pandas dataframe
url_working_data["Image index"] = index
```

```python
# Move image index column to first column
cols = list(url_working_data.columns)
cols = [cols[-1]] + cols[:-1]
url_working_data = url_working_data[cols]
```

```python
url_working_data
```

```python
# Clean category columns
url_working_data['main category'] = url_working_data['main category'].str.replace('\t','')
url_working_data['Cat option 1'] = url_working_data['Cat option 1'].str.replace('\t','')
url_working_data['Cat option 2'] = url_working_data['Cat option 2'].str.replace('\t','')
url_working_data['Cat option 3'] = url_working_data['Cat option 3'].str.replace('\t','')
url_working_data['Cat option 4'] = url_working_data['Cat option 4'].str.replace('\t','')
```

```python
# Replacing NaN with empty string
import numpy as np
url_working_data['main category'] = url_working_data['main category'].replace(np.nan, '', regex=True)
url_working_data['Cat option 1'] = url_working_data['Cat option 1'].replace(np.nan, '', regex=True)
url_working_data['Cat option 2'] = url_working_data['Cat option 2'].replace(np.nan, '', regex=True)
url_working_data['Cat option 3'] = url_working_data['Cat option 3'].replace(np.nan, '', regex=True)
url_working_data['Cat option 4'] = url_working_data['Cat option 4'].replace(np.nan, '', regex=True)
```

```python
# Use one hot encoding method to create vectors
```

```python
# Combining categories columns into one
#data['categories'] = data[['main category', 'Cat option 1', 'Cat option 2', 'Cat option 3', 'Cat option 4']].values.tolist()
url_working_data['categories'] = url_working_data.apply(lambda x: set([x['main category'],
                                        x['Cat option 1'],
                                        x['Cat option 2'],
                                        x['Cat option 3'],
                                        x['Cat option 4']]),axis=1) 
```

```python
# One hot encoding
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
one_hot_data = pd.DataFrame(mlb.fit_transform(url_working_data['categories']),columns=mlb.classes_)
```

```python
# Drop first column from dataframe
del one_hot_data['']
```

```python
one_hot_data
```

```python
# Adding image index and url to one hot encoded dataframe
```

```python
one_hot_data['Image index'] = url_working_data['Image index'].tolist()
one_hot_data['Image URL'] = url_working_data['Image URL'].tolist()
```

```python
one_hot_data
```

```python
# Move image index column to first column
cols = list(one_hot_data.columns)
cols = [cols[-2]] + cols[:-2] + [cols[-1]]
one_hot_data = one_hot_data[cols]
```

```python
# Download input images
image_urls = one_hot_data['Image URL']
image_index = 0
```

```python
import requests

def download_image(image_name,image_url):
    with open('data/' + image_name + '.jpeg', 'wb') as handle:
        response = requests.get(image_url, stream=True)

        if not response.ok:
            print(response)

        for block in response.iter_content(1024):
            if not block:
                break

            handle.write(block)
```

```python
def download_images(image_urls,image_index):
    for image_url in image_urls:
        # Download the image
        download_image(str(image_index),image_url)
    
        # Increment index
        image_index = image_index + 1
```

```python
download_images(image_urls,image_index)
```
